## **Root Finding**
---

>### Using NLsolve to find the root
- **Goal**: Find the intersection of 
$$
\begin{aligned}
Circle: f1(x1, x2) = x1^2 + x2^2 - r1^2 \\ 
Hyperbola: f2(x1, x2) = x1^2 - x2^2 - r2^2
\end{aligned}
$$
- **Key Takeaway**: 
    - define system of equations 
    - initial guess
    - nlsolve
    - extract solution

Packages used: 
<span style="color:yellow; background-color:green">NLsolve, Plots</span>

```julia
using NLsolve, Plots
include("src\\plot_library.jl")
```

```julia
# Define variables and system of equations
r1, r2 = sqrt(2), sqrt(0.05)
function system_of_equations!(F,x)      # RETURN IN PLACE
    F[1] = x[1]^2 + x[2]^2 - r1^2
    F[2] = x[1]^2 - x[2]^2 - r2^2
end

# Initial guess and solve
initial_guess = [1.3, 0.6]
result = nlsolve(system_of_equations!, initial_guess)

# Extract solution
root = result.zero

# Print the root to verify
println("Root found by NLsolve: ", root)
```
